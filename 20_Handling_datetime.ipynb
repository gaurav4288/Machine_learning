```python
# converting to date-time datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extract year
date['date_year']=date['date'].dt.year

# Extract month
date['date_month_no']=date['date'].dt.month
date['date_month_name']=date['date'].dt.month_name()

# Extract Days
date['date_day']=date['date'].dt.day

# Extract day of week
date['date_dow']=date['date'].dt.day_of_week
date['date_dow_name']=date['date'].dt.day_name()

# date isweekend
date['date_is_weekend'] = np.where(date['date_dow'].isin([5,6]),1,0)

# Extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week

# Extract quarter
date['quarter'] = date['date'].dt.quarter

# Extract semester
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)


date.drop(columns=['product_id','city_id','orders'])

date.head()
```

```python
# Extract time elapsed b/w dates

import pytz
import datetime

# Localize the 'date' column to IST if it is not already timezone-aware
ist = pytz.timezone('Asia/Kolkata')
date['date'] = date['date'].dt.tz_localize(ist)

# Get the current time in IST
utc_now = datetime.datetime.now(datetime.timezone.utc)
ist_now = utc_now.astimezone(pytz.timezone('Asia/Kolkata'))

# Calculate the time elapsed
date['time_elapsed'] = ist_now - date['date']

date.head()
```

```python
(ist_now-date['date']).dt.days
```

```python
# Months passed
np.round((ist_now-date['date'])/np.timedelta64(1,'M'),0)
```
