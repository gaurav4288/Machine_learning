```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# df['Survived'].value_counts().plot(kind='bar')
sns.countplot(x='Survived', hue='Survived', data=df, palette=["blue", "orange"], legend=False)
df['Survived'].value_counts()
```

```python
sns.countplot(x='Sex',data=df)
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], edgecolor='black',bins=8)
plt.show()
```

```python
sns.displot(x='Age',data=df,stat='probability',kde=True)
```

```python
sns.boxplot(data=df,x='Fare')
```
